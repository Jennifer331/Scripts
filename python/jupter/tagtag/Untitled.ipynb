```python
def extract_feature(p1, r1, p2, r2):
    t = 10 ** ((r1 - r2) / 40)
    for k1 in range(12):
        for k2 in (k1, k1 + 2):
            m = (2*np.pi*(t*k1-k2) + (t*p1-p2))/(t - 1)
            print(k1, k2, t, m)
            plt.plot(m)
```

```python
df_water30 = df_water.groupby('DISTANCE').get_group(30)
df_water33 = df_water.groupby('DISTANCE').get_group(33)
extract_feature(df_water30['PHASE'], df_water30['RSSI'], df_water33['PHASE'], df_water33['RSSI'])
```
