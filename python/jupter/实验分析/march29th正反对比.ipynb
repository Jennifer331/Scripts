```python
def plot_raw(f1, f2, epc, attr, folder1='D:\\Atom\\exp\\20210329\\open_grill', folder2='D:\\Atom\\exp\\20210329\\open_grill'):
    df_v_f = pd.read_csv(os.path.join(folder1, f1)).groupby('EPC').get_group(epc)
    df_v_t = pd.read_csv(os.path.join(folder2, f2)).groupby('EPC').get_group(epc)
    plt.scatter(df_v_f['CHANNEL'], df_v_f[attr])
    plt.scatter(df_v_t['CHANNEL'], df_v_t[attr])
    plt.legend(['front', 'end'])
    plt.show()
```

```python
folder = 'D:\\Atom\\exp\\20210328\\side'
df_v_f = pd.read_csv(os.path.join(folder, '100cm_front_oil.csv')).groupby('EPC').get_group('3008 33B2 DDD9 0140 0000 0040')
df_v_t = pd.read_csv(os.path.join(folder, '100cm_tail_oil.csv')).groupby('EPC').get_group('3008 33B2 DDD9 0140 0000 0040')
plt.scatter(df_v_f['CHANNEL'], df_v_f['PHASE'])
plt.scatter(df_v_t['CHANNEL'], df_v_t['PHASE'])
plt.legend(['front', 'end'])
plt.title('oil')
plt.plot()
```
